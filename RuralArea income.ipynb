```python
#Identifying Outliers

cols=['Median_HH_Inc_ACS', 'PerCapitaInc',
       'Poverty_Rate_0_17_ACS', 'Poverty_Rate_ACS', 'Deep_Pov_All',
       'Deep_Pov_Children', 'NumAll_inPOV_ACS', 'PCTPOV017', 'POV017',
       'MedHHInc', 'POVALL', 'PCTPOVALL', 'Num_inPOV_0_17_ACS']

outlier_info=[]

for column in cols:
    q1=income[column].quantile(0.25)
    q3=income[column].quantile(0.75)
    iqr=q3-q1 
    ul= q3+1.5*iqr 
    ll=q1-1.5*iqr
    outlier=income[(income[column] <ll) | (income[column] >ul)]
    outlier_info.append({'Feature': column, 'No. of Outliers': outlier.shape[0]})

outlier_df = pd.DataFrame(outlier_info)
outlier_df['Outlier Proportion']=outlier_df['No. of Outliers']*100/income.shape[0]
outlier_df
```

```python
print(max(outlier_df['Outlier Proportion']))
```

```python
outlier_desc=outlier_df.sort_values(by='Outlier Proportion', ascending=False)[['Feature', 'Outlier Proportion']]
print(outlier_desc)
```

```python
#drop rows containing Null values 
before=income.shape[0]
cols=['Median_HH_Inc_ACS', 'PerCapitaInc',
       'Poverty_Rate_0_17_ACS', 'Poverty_Rate_ACS', 'Deep_Pov_All',
       'Deep_Pov_Children', 'NumAll_inPOV_ACS', 'PCTPOV017', 'POV017',
       'MedHHInc', 'POVALL', 'PCTPOVALL', 'Num_inPOV_0_17_ACS']

for column in cols:  
    income = income.dropna(subset=[column])

after=income.shape[0]
print('Number of rows with null values removed:', before-after)
```
