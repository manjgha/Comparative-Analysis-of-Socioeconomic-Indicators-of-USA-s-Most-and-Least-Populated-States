```python
#Identifying Outliers

cols=['Pop_change_Rate_2020_2021', 'POPESTIMATE2021',
       'Net_Migration_Rate_2020_2021', 'Natural_Change_Rate_2020_2021',
       'Net_InterMigrationRate_2020_2021', 'PopChangeRate1020',
       'PopDensity2020', 'Under18Pct2020', 'Age65AndOlderPct2020',
       'WhiteNonHispanicPct2020', 'BlackNonHispanicPct2020',
       'AsianNonHispanicPct2020', 'NativeAmericanNonHispanicPct2020',
       'HispanicPct2020', 'MultipleRacePct2020',
       'NonHispWhitePopChangeRate1020', 'NonHispBlackPopChangeRate1020',
       'NonHispAsianPopChangeRate1020', 'NonHispNatAmerPopChangeRate1020',
       'HispanicPopChangeRate1020', 'MultipleRacePopChangeRate1020',
       'WhiteNonHispanicNum2020', 'BlackNonHispanicNum2020',
       'AsianNonHispanicNum2020', 'NativeAmericanNonHispanicNum2020',
       'HispanicNum2020', 'MultipleRaceNum2020', 'ForeignBornPct',
       'ForeignBornEuropePct', 'ForeignBornMexPct', 'NonEnglishHHPct',
       'Ed1LessThanHSPct', 'Ed2HSDiplomaOnlyPct', 'Ed3SomeCollegePct',
       'Ed4AssocDegreePct', 'Ed5CollegePlusPct', 'AvgHHSize', 'FemaleHHPct',
       'HH65PlusAlonePct', 'OwnHomePct', 'Ed3SomeCollegeNum',
       'Ed2HSDiplomaOnlyNum', 'Ed1LessThanHSNum', 'TotalPopACS',
       'ForeignBornNum', 'TotalPop25Plus', 'ForeignBornEuropeNum',
       'ForeignBornCaribNum', 'ForeignBornMexNum', 'Ed5CollegePlusNum',
       'HH65PlusAloneNum', 'Ed4AssocDegreeNum', 'ForeignBornAfricaNum',
       'ForeignBornAsiaNum', 'NonEnglishHHNum', 'OwnHomeNum', 'FemaleHHNum',
       'TotalHH', 'ForeignBornCentralSouthAmPct',
       'ForeignBornCentralSouthAmNum', 'ForeignBornCaribPct',
       'ForeignBornAsiaPct', 'TotalOccHU', 'ForeignBornAfricaPct',
       'Age65AndOlderNum2020', 'Under18Num2020', 'LandAreaSQMiles2020',
       'TotalPop2020', 'NaturalChange1019', 'TotalPopEst2015',
       'TotalPopEst2016', 'TotalPopEst2017', 'TotalPopEst2018',
       'Net_International_Migration_2010_2019', 'TotalPopEst2019',
       'TotalPopEst2012', 'NetMigration1019',
       'Net_International_Migration_Rate_2010_2019', 'NaturalChangeRate1019',
       'NetMigrationRate1019', 'PopChangeRate1019', 'TotalPopEst2014',
       'TotalPopEst2013', 'TotalPopEstBase2010', 'TotalPopEst2011',
       'TotalPopEst2010', 'ESTIMATESBASE2020', 'POPESTIMATE2020',
       'NaturalChg2020_2021', 'Net_InterMigration_2020_2021',
       'Net_Migration_2020_2021']

outlier_info=[]

for column in cols:
    q1=people[column].quantile(0.25)
    q3=people[column].quantile(0.75)
    iqr=q3-q1 
    ul= q3+1.5*iqr 
    ll=q1-1.5*iqr
    outlier=people[(people[column] <ll) | (people[column] >ul)]
    outlier_info.append({'Feature': column, 'No. of Outliers': outlier.shape[0]})

outlier_df = pd.DataFrame(outlier_info)
outlier_df['Outlier Proportion']=outlier_df['No. of Outliers']*100/people.shape[0]
outlier_df
```
